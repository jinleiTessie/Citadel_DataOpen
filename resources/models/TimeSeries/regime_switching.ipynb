```python
import pandas as pd
import statsmodels.api as sm
# Get the RGNP data to replicate Hamilton
from statsmodels.tsa.regime_switching.tests.test_markov_autoregression import rgnp
dta_hamilton = pd.Series(rgnp, index=pd.date_range('1951-04-01', '1984-10-01', freq='QS'))

# Plot the data
dta_hamilton.plot(title='Growth rate of Real GNP', figsize=(12,3))

# Fit the model
mod_hamilton = sm.tsa.MarkovAutoregression(dta_hamilton, k_regimes=2, order=4, trend='c', switching_ar=False, switching_trend=False, switching_variance=False)
# k_regimes nb of regimes, order the order of the autoregressive lag polynomial,
# trend can be nc, c, t, ct (no, constant, time, both)
#switching_ar to know whether coeffs change or not. Provide an iterable to change only some.
#
res_hamilton = mod_hamilton.fit()
```
